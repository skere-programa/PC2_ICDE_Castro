```python
import pandas as pd
#Primero importamos el Excel que se va a limpiar
dataframe = pd.read_excel('https://github.com/skere-programa/PC2_ICDE_Christian_Castro/blob/main/HDI.xlsx?raw=true')
dataframe
```

```python
#Podemos ver que existen elementos non null, por lo que será necesario hacer una limpieza de datos
dataframe.info()
```

```python
#Como las 3 primeras filas no tienen datos, vamos a eliminarlas
limpiar_dataframe = dataframe.iloc[3:]
limpiar_dataframe
```

```python
#Se procederá a eliminar las columnas que tienen nombres NaN pero se conservarán las 2 primeras columnas
cols_to_keep = limpiar_dataframe.columns[:2]
next_dataframe = pd.concat([limpiar_dataframe[cols_to_keep], limpiar_dataframe.loc[:, limpiar_dataframe.iloc[0].notna()].iloc[:, 2:]], axis=1)
next_dataframe
```

```python
#Se eliminará todas las filas que sus valores sean NaN, excepto en la segunda columna
#En el código se busca que desde la tercera columna en adelante si tienen valores vacíos, devuelve True
dframe_depurado = next_dataframe[~(next_dataframe.iloc[:, 1].notna() & next_dataframe.iloc[:, 2:].isna().all(axis=1))]
dframe_depurado
```

```python
#Una vez ya seleccionamos las columnas,ahora se eliminará las filas que solo tengan valores de NaN (vacíos)
dframe_depurado = dframe_depurado.dropna(how='all')
dframe_depurado
```

```python
#Podemos notar que se eliminaron 10 columnas
#Ahora, las celdas que tienen NaN se reemplazarán con espacios en blanco
dframe_depurado = dframe_depurado.fillna('')
dframe_depurado
```

```python
#Ya se reemplazaron los NaN; sin embargo, aún falta cambiar los símbolos ".." y "-" con espacios en blanco ' '
df_depurado_último = dframe_depurado.replace({'..': '', '—': ''})
df_depurado_último
```
